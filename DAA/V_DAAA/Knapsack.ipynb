```python
# Knapsack by minimun weight

def knap_minWeight(l1,capacity):
    profit=0
#     w1=sortWeight(l1)
    w1 = sorted(l1, key=lambda x: x[1])
    for i in range(len(l1)):
        if(capacity>0 and w1[i][1]<=capacity):
            capacity-=w1[i][1]
            profit+=w1[i][0]
        elif capacity>0:
            profit+=(w1[i][0]*capacity)/w1[i][1]
            capacity=0
        else:
            break
    return profit
```

```python
start1=time.perf_counter()
profit_weight=knap_minWeight(l1,capacity)
print(profit_weight)
end1=time.perf_counter()
t1=end1-start1
```

```python
# Knapsack for maximum profit

def knap_maxProfit(l1,capacity):
    profit=0
#     w1=sortProfit(l1)
    w1 = sorted(l1, key=lambda x: x[0], reverse=True)
    for i in range(len(l1)):
        if(capacity>0 and w1[i][1]<=capacity):
            capacity-=w1[i][1]
            profit+=w1[i][0]
        elif capacity>0:
            profit+=(w1[i][0]*capacity)/w1[i][1]
            capacity=0
        else:
            break
    return profit
```

```python
start2=time.perf_counter()
profit_weight=knap_maxProfit(l1,capacity)
print(profit_weight)
end2=time.perf_counter()
t2=end2-start2
```

```python
# Knapsack for maximum profit/ weight ratio

def knap_maxProfitWeight(l1,capacity):
    profit=0
#     w1=sortPW(l1)
    w1 = sorted(l1, key=lambda x: x[2], reverse=True)
    for i in range(len(l1)):
        if(capacity>0 and w1[i][1]<=capacity):
            capacity-=w1[i][1]
            profit+=w1[i][0]
        elif capacity>0:
            profit+=(w1[i][0]*capacity)/w1[i][1]
            capacity=0
        else:
            break
    return profit
```

```python
start3=time.perf_counter()
profit_weight=knap_maxProfitWeight(l1,capacity)
print(profit_weight)
end3=time.perf_counter()
t3=end3-start3
```
